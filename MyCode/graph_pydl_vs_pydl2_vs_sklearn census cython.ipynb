```python
def graphh(axe1, axe2, x_size, y_size, nameX=None, nameY=None, f=None):
    legends = ['< 0', '< 0.1', '< 0.2', '< 0.3', 'no relab']
    
    list_k2 = list_k.copy()
    for k in range(0, len(list_k)):
        list_k2[k] = "k="+str(list_k[k])
    for depth in df_sklearn['depth'].unique():
        fig, axes = plt.subplots()
        index_legends = 0
        y_values = list()
        x_values = list()
        y_values_pydl = list()
        x_values_pydl = list()
        list_point=list()
        for x_axe, y_axe in axe1:

            tmp = df_sklearn.loc[(df_sklearn["k"] == 0) & (df_sklearn["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                list_point.append(plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS8[index_legends], marker='x'))
            else:
                list_point.append(plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS8[index_legends]))

            tmp = df_pydl.loc[(df_pydl["k"] == 0) & (df_pydl["depth"] == depth)]
            x_values_pydl.append(tmp[x_axe].mean())
            y_values_pydl.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS8[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS8[index_legends])

            index_legends +=1
        plt.plot(x_values, y_values, label="Sklearn (relab)", c=_COLORS3[0], linestyle="-")
        plt.plot(x_values_pydl, y_values_pydl, label="DL8.5 (relab)", c=_COLORS3[1], linestyle="-")

        """
        y_values = list()
        x_values = list()
        x_axe = axe2[0]
        y_axe = axe2[1]
        index_legends = 0
        #for k in [0, 1, 5, 10, 50, 100, 250, 500, 1000, 5000, 100000]:
        for k in [0, 5, 10, 100, 500, 1000, 5000]:
            tmp = df_pydl.loc[(df_pydl["k"] == k) & (df_pydl["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends])

            index_legends +=1

        plt.plot(x_values, y_values, label="Pydl (constantes)", c=colors4[2])
        """
        # PYDL 2
        y_values = list()
        x_values = list()
        x_axe = axe2[0]
        y_axe = axe2[1]
        index_legends = 0
        #for k in [0, 1, 5, 10, 50, 100, 250, 500, 1000, 5000, 100000]:
        for k in list_k:
            tmp = df_pydl2.loc[(df_pydl2["k"] == k) & (df_pydl2["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS11[index_legends], marker='v')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS11[index_legends], marker='^')

            index_legends +=1

        plt.plot(x_values, y_values, label="DL8.5 (DATC approach 2)", c=_COLORS3[2], linestyle="-")
        
        
        # PYDL 2
        y_values = list()
        x_values = list()
        x_axe = axe2[0]
        y_axe = axe2[1]
        index_legends = 0
        
        #for k in [0, 1, 5, 10, 50, 100, 250, 500, 1000, 5000, 100000]:
        for k in list_k:
            tmp = df_pydl5.loc[(df_pydl2["k"] == k) & (df_pydl5["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                list_point.append(plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS11[index_legends], marker='v'))
            else:
                list_point.append(plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=_COLORS11[index_legends], marker='^'))

            index_legends +=1

        plt.plot(x_values, y_values, label="DL8.5 (DATC approach 3)", c=_COLORS3[3], linestyle="-")
        
        
        plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
        plt.title(f'Depth {depth}')
        plt.ylabel('discrimination')
        plt.xlabel('accuracy')
        plt.ylim(y_size[0], y_size[1])
        plt.xlim(x_size[0], x_size[1])
        
        if nameX==None:
            plt.xlabel(x_axe)
        else:
            plt.xlabel(nameX)
        if nameY==None:
            plt.ylabel(y_axe)
        else:
            plt.ylabel(nameY)
        
        l1 = plt.legend()
        list_kk = ['no relab', 'threshold 0.3', 'threshold 0.2',  'threshold 0.1', 'threshold 0'] +list_k2
        plt.legend(list_point, list_kk, loc='upper left', scatterpoints=1)
        plt.gca().add_artist(l1)
        
        if f != None:
            plt.savefig('plots_census/'+f+'_depth_'+str(depth)+'.svg',dpi=350)
        
        


        plt.show()
```
