```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('max_columns', None)
FILE_NAME = "lawsuit_k_supp2"

df_pydl = pandas.read_csv("lawsuit/save/"+FILE_NAME+"_relab.csv")
df_pydl2 = pandas.read_csv("lawsuit2/save/"+FILE_NAME+".csv")
df_pydl3 = pandas.read_csv("lawsuit3/save/"+FILE_NAME+".csv")
df_sklearn = pandas.read_csv("lawsuit/save/"+FILE_NAME+"_relab_sklearn.csv")
```

```python
_COLORS4 = ['#ff0000', '#00ff00', '#0000ff', '#87cefa']
_COLORS5 = ['#008080', '#ffa500', '#00ff00', '#0000ff', '#ff1493']
_COLORS7 = ['#2e8b57', '#ff0000', '#ffd700', '#c71585', '#00ff00', '#0000ff', '#1e90ff']
_COLORS10 = ['#006400', '#00008b', '#b03060', '#ff4500', '#ffd700', '#7cfc00', '#00ffff', '#ff00ff', '#6495ed',
          '#ffdab9']
_COLORS11 = ['#8b4513', '#006400', '#4682b4', '#4b0082', '#ff0000', '#ffff00', '#00ff7f', '#00ffff', '#ff00ff', '#ffff54',
             '#ffe4c4']
```

```python
def graphh(axe1, axe2, x_size, y_size):
    colors11 = _COLORS11.copy()
    colors5 = _COLORS11.copy()
    colors4 = _COLORS5.copy()
    legends = ['< 0', '< 0.1', '< 0.2', '< 0.3', 'no relab']
    fig, axes = plt.subplots()

    for depth in df_sklearn['depth'].unique():
        index_legends = 0
        y_values = list()
        x_values = list()
        y_values_pydl = list()
        x_values_pydl = list()
        for x_axe, y_axe in axe1:

            tmp = df_sklearn.loc[(df_sklearn["k"] == 0) & (df_sklearn["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors5[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors5[index_legends])

            tmp = df_pydl.loc[(df_pydl["k"] == 0) & (df_pydl["depth"] == depth)]
            x_values_pydl.append(tmp[x_axe].mean())
            y_values_pydl.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors5[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors5[index_legends])

            index_legends +=1
        plt.plot(x_values, y_values, label="Sklearn", c=colors4[0])
        plt.plot(x_values_pydl, y_values_pydl, label="Pydl (relab)", c=colors4[1])

        y_values = list()
        x_values = list()
        x_axe = axe2[0]
        y_axe = axe2[1]
        index_legends = 0
        #for k in [0, 1, 5, 10, 50, 100, 250, 500, 1000, 5000, 100000]:
        for k in [0, 5, 10, 100, 500, 1000, 5000]:
            tmp = df_pydl.loc[(df_pydl["k"] == k) & (df_pydl["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends])

            index_legends +=1

        plt.plot(x_values, y_values, label="Pydl (constantes)", c=colors4[2])

        # PYDL 2
        y_values = list()
        x_values = list()
        x_axe = axe2[0]
        y_axe = axe2[1]
        index_legends = 0
        #for k in [0, 1, 5, 10, 50, 100, 250, 500, 1000, 5000, 100000]:
        for k in [0, 5, 10, 100, 500, 1000, 5000]:
            tmp = df_pydl2.loc[(df_pydl2["k"] == k) & (df_pydl2["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends])

            index_legends +=1

        plt.plot(x_values, y_values, label="Pydl (constantes2)", c=colors4[3])

        # PYDL 3
        y_values = list()
        x_values = list()
        x_axe = axe2[0]
        y_axe = axe2[1]
        index_legends = 0
        #for k in [0, 1, 5, 10, 50, 100, 250, 500, 1000, 5000, 100000]:
        for k in [0, 5, 10, 100, 500, 1000, 5000]:
            tmp = df_pydl3.loc[(df_pydl3["k"] == k) & (df_pydl3["depth"] == depth)]
            x_values.append(tmp[x_axe].mean())
            y_values.append(tmp[y_axe].mean())

            if index_legends == 0:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends], marker='x')
            else:
                plt.scatter(tmp[x_axe].mean(), tmp[y_axe].mean(), c=colors11[index_legends])

            index_legends +=1

        plt.plot(x_values, y_values, label="Pydl (constantes3)", c=colors4[4])

        plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
        plt.title(f'Depth {depth}')
        plt.ylabel('discrimination')
        plt.xlabel('accuracy')
        plt.ylim(y_size[0], y_size[1])
        plt.xlim(x_size[0], x_size[1])
        plt.legend()

        plt.show()
```
